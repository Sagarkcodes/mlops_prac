```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Generate synthetic time series data (sine wave)
data = np.sin(np.arange(100) * 0.1)  # 100 time steps

# Prepare data (past 10 values  predict next value)
X, y = [], []
for i in range(len(data) - 10):
    X.append(data[i:i+10])
    y.append(data[i+10])

X, y = np.array(X), np.array(y)
X = X.reshape(-1, 10, 1)  # Reshape for LSTM (samples, timesteps, features)

# Define simple LSTM model
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(20, input_shape=(10, 1)),
    tf.keras.layers.Dense(1)  # Output next value
])

model.compile(optimizer="adam", loss="mse")
model.fit(X, y, epochs=10, verbose=0)  # Train model

# Predict next 10 values
preds = []

# Start with the last sequence in the training data
seq = X[-1]

for _ in range(10):
    pred = model.predict(seq.reshape(1, 10, 1), verbose=0)[0, 0]
    preds.append(pred)
    
    # Shift sequence by removing the first element and adding the predicted value
    seq = np.roll(seq, -1)
    seq[-1] = pred  # Update last value with prediction

# Plot results
plt.plot(data, label="Actual")
plt.plot(range(90, 100), preds, 'ro-', label="Predicted")
plt.legend()
plt.show()
```
